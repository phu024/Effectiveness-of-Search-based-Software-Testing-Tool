Compare the average of EvoSuite-60s, UTBot-60s Evosuite-120s, UTBot-120s, Evosuite-300s and UTBot-300s

```python
compare = pd.concat([statements['Sample CUTs'],Evosuite_60s_avg,UTBot_60s_avg,Evosuite_120s_avg,UTBot_120s_avg,Evosuite_300s_avg,UTBot_300s_avg],axis=1)
```

```python
compare
```

```python
compare.describe()
```

------------

```python
statementsCoverage = pd.melt(compare, id_vars=['Sample CUTs'], value_vars=['Evosuite-60s_Mean','UTBot-60s_Mean','Evosuite-120s_Mean','UTBot-120s_Mean','Evosuite-300s_Mean','UTBot-300s_Mean'], var_name='Tool', value_name='Statements Coverage')
```

```python
statementsCoverage['Time(s)'] = statementsCoverage['Tool'].str.extract('(\d+)', expand=False)
```

```python
statementsCoverage['Tool'] = statementsCoverage['Tool'].str.replace('-\d+s_Mean','')
```

```python
statementsCoverage
```

```python
ax = sns.boxplot(x="Time(s)", y="Statements Coverage", hue="Tool", data=statementsCoverage, palette="Set3",)
sns.move_legend(ax, "lower right",bbox_to_anchor=(.5, 1), ncol=3, title=None, frameon=False,)
```

```python
sns.violinplot(x="Time(s)", y="Statements Coverage", hue="Tool", data=statementsCoverage, palette="Set3")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0,title='Tools')
```

```python
sns.boxenplot(x="Time(s)", y="Statements Coverage", hue="Tool", data=statementsCoverage, palette="Set3")
```
